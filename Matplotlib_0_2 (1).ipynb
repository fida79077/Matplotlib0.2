```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/python/LIBRARIES/MATPLOTLIB/company_sales_data.csv')

profitlist = df ['total_profit'].tolist()
#labels = ['low','average','good','']

profit_range = [150000,175000,200000,225000,300000,350000]

plt.hist(profitlist,profit_range,rwidth=0.60,label = 'profit of company' , color = 'm' )

plt.xlabel('profit range in dollar')
plt.ylabel('actual profit in dollar')
#plt.ystick(profitlist)
plt.legend(loc = 'upper left')
plt.xticks(profit_range)
plt.title('profit data')

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
month = np.arange(1,13)   # we can not add numbers to list. so want to convert it to array

death21 = [12,3,23,5,27,7,12,4,14,9,16,10]
death22 = [4,3,28,14,29,8,10,30,20,29,17,25]

plt.title('monthly death rate of corona patients ')

plt.ylim(1,30)

plt.xlabel('months')
plt.ylabel('death rate')

x = range(1,13)
y = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']

plt.xticks(x,y,rotation = 90)

plt.figure(figsize=(6,5))

plt.bar(month,death21,color='c' ,label = '2021',width = 0.3)
plt.bar(month + 0.4,death22,color = 'm' , label = '2022',width = 0.3)
plt.legend(loc=1)      #loc=1,upperryt,2 means upper left ,0 means best position , 3 means lower left , 4 means lower right ,10 means centre

plt.text(x = 4.5,y = 25,s = 'corona')
plt.grid(True , color = 'w')


#plt.style.available
#plt.style.use('ggplot')
#plt.style.use('_mpl-gallery-nogrid')
#plt.style.use('seaborn-v0_8-paper')
plt.style.use( 'dark_background')

plt.show()
```
